# Relatorio de analise VI

```python
# criando novas variaveis
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
dados.head()
```

```python
dados['Valor_Bruto'] = dados['Valor'] + dados['Condominio']
```

```python
dados.head()
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head()
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados.head()
```

```python
dados['Valor Bruto m2'] = (dados['Valor_Bruto'] / dados['Area']).round(2)
```

```python
dados.head()
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
casa
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

## Excluindo variaveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor_Bruto', 'Valor Bruto m2']])
dados_aux.head()
```

```python
del dados_aux['Valor_Bruto']
```

```python
dados_aux.head()
```

```python
dados_aux.pop('Valor Bruto m2')
```

```python
dados_aux.head()
```

```python
dados.drop(['Valor m2', 'Valor Bruto m2'], axis=1, inplace=True)
dados.head()
```

```python
dados.to_csv('dados/aluguel_residencial.csv', sep=';', index=False)
```
