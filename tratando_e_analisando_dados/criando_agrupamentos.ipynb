```python
import pandas as pd

dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados.head()
```

```python
dados.Valor.mean()
```

```python
dados.Bairro.unique()
```

```python
bairros = ['Tijuca', 'Olaria', 'Leblon', 'Ipanema',
       'Campo Grande', 'Recreio dos Bandeirantes', 'Jardim Oceânico',
       'Humaitá', 'Península', 'Méier', 'Vargem Pequena', 'Maracanã',
       'Jacarepaguá', 'São Conrado', 'Vila Valqueire', 'Gávea',
       ]

selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
dados
```

```python
grupo_bairro =  dados.groupby('Bairro')
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
for bairro, data in grupo_bairro:
    print('{} -> R$ {}'.format(bairro, data.Valor.mean().round(2)))
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

## Estatisticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
```
